### Series

Una Serie es una estructura de datos unidimensional etiquetada en [Pandas](https://pandas.pydata.org/). Puede contener cualquier tipo de dato y es similar a una columna en una hoja de clculo o un diccionario. Cada elemento en una `Serie` est asociado con una etiqueta nica llamada ndice. La Serie proporciona capacidades de indexacin y manipulacin de datos poderosas.

Se puede crear a partir de una lista o array de la siguiente manera:

````python

# Ejemplo de creacin de una Serie
import pandas as pd
s = pd.Series([1, 3, 5, None, 6, 8])
s
````

Una `Serie` de [Pandas](https://pandas.pydata.org/) se comporta de muchas maneras como un array de NumPy y en otras como un diccionario estndar de Python. Si tenemos una Serie `s`, podemos seleccionar datos usando la indexacin de array (usando un ndice explcito) y la indexacin de diccionario (usando el ndice asignado).


````python
s = pd.Series([1, 3, 5, None, 6, 8], index=['a', 'b', 'c', 'd', 'e', 'f'])
print(s['b'])  # indexacin de diccionario
print(s[1])  # indexacin de array
````


### Index

En la seccin anterior vimos que las `Series` contienen un ndice explcito que permite hacer referencia y modificar los datos, esto ser tambin cierto para los `DataFrames`. Este ndice es una estructura similar a un array inmutable en `Pandas` que contiene las etiquetas de las filas o columnas en una `Serie` o `DataFrame`. Proporciona as, una forma de identificar de manera nica cada fila o columna y permite una recuperacin eficiente y alineacin de datos. El ndice puede contener etiquetas de cualquier tipo de dato que sea hashable, como enteros, caracteres o fechas.

````python
# Ejemplo de creacin de un objeto Index

index = pd.Index([0,1,"hola"])
obj2 = pd.Series([1.5, -2.5, 0], index=index)
obj2.index
````

Los objetos Index son inmutables y no pueden ser modificados por el usuario. La inmutabilidad hace que sea ms seguro compartir los objetos Index entre mltiples `DataFrames` y `arrays`, sin el riesgo de que el ndice sea modificado.

### Manipulando datos

#### Filtrado de filas y columnas

El filtrado de datos es una de las operaciones ms comunes que se realizan en los conjuntos de datos. [Pandas](https://pandas.pydata.org/) proporciona varias formas de filtrar filas y columnas.

Veamos algunos ejemplos:

````python
# Filtrado de columnas
# Para filtrar columnas, podemos simplemente seleccionar las columnas que queremos mantener.
# Por ejemplo, si queremos mantener solo las columnas 'Ciudad' y 'Alimentos y bebidas no alcohlicas', podemos hacer lo siguiente:
filtered_df = df[['Ciudad', 'Alimentos y bebidas no alcohlicas']]
filtered_df.head()
````

### Combinando  `DataFrames`

[Pandas](https://pandas.pydata.org/) proporciona varias formas de combinar `DataFrames`, incluyendo `merge` y `join`.

- `merge` se utiliza para combinar dos `DataFrames` a lo largo de una columna en comn.
- `join` se utiliza para combinar dos `DataFrames` a lo largo de su ndice.


Vamos a ver algunos ejemplos de cmo usar `merge` y `join`. Para ello, primero dividamos la ENPH en dos:
    
- `df1` que contiene las columnas: `Ciudad`, `Alimentos y bebidas no alcohlicas`, `Bebidas alcohlicas y tabaco`, `Prendas de vestir y calzado`, A`lojamiento, agua, electricidad, gas y otros combustibles`, `Muebles, artculos para el hogar y para la conservacin ordinaria del hogar`
- `df2` que contiene `Ciudad`, `Salud`, `Transporte`, `Informacin y comunicacin`, `Recreacin y cultura`, `Educacin`, `Restaurantes y hoteles`, `Bienes y servicios diversos`.


````python
# Dividimos los datos en dos DataFrames
df1 = df[['Ciudad', 'Alimentos y bebidas no alcohlicas', 'Bebidas alcohlicas y tabaco', 'Prendas de vestir y calzado', 'Alojamiento, agua, electricidad, gas y otros combustibles', 'Muebles, artculos para el hogar y para la conservacin ordinaria del hogar']]
df2 = df[['Ciudad', 'Salud', 'Transporte', 'Informacin y comunicacin', 'Recreacin y cultura', 'Educacin', 'Restaurantes y hoteles', 'Bienes y servicios diversos']]
````

#### Merge de DataFrames

La funcin `merge()` se utiliza para combinar dos (o ms) `DataFrames` en base a una columna comn (clave).

````python
# Merge df1 y df2 on 'Ciudad'
merge_df = pd.merge(df1, df2, on='Ciudad')

# Las primeras filas del nuevo DataFrame
merge_df.head()
````

#### Join de DataFrames

La funcin `join()` se utiliza para combinar columnas de dos `DataFrames` potencialmente diferentes en un solo `DataFrame`.

````python
# Fijamos 'Ciudad' como el indice de ambos DataFrames
df1.set_index('Ciudad', inplace=True)
df2.set_index('Ciudad', inplace=True)

# Unimos df1 y df2
join_df = df1.join(df2, lsuffix='_df1', rsuffix='_df2')

# Las primeras filas del nuevo DataFrame
join_df.head()
````

### Agrupando y Agregando

`Pandas` proporciona mtodos de agrupacin y agregacin flexibles, lo que permite dividir los datos en grupos y luego aplicar funciones de agregacin a cada grupo para realizar anlisis de datos complejos.

Para ilustrar esto, vamos a crear una nueva variable que indique si una ciudad es una de las [ciudades en Colombia con ms de dos millones de habitantes](https://es.wikipedia.org/wiki/Anexo:Municipios_de_Colombia_por_poblaci%C3%B3n) (Bogot, Medelln y Cali) o no. Luego, vamos a agrupar los datos por esta nueva variable y calcular la suma de los gastos en Salud, Transporte y Educacin para cada grupo.

Podemos crear esta nueva variable en el `DataFrame` utilizando la funcin `loc()`. En el siguiente ejemplo, vamos a crear una variable llamada 'Mas_2_mill' que ser 'Si' si la ciudad es Bogot, Medelln, o Cali, y 'No' en caso contrario.

````python
# Creamos la variable 'Mas_2_mill' usando loc
df.loc[df['Ciudad'].isin(['Bogot', 'Medelln', 'Cali']), 'Mas_2_mill'] = 'Si'
df.loc[~df['Ciudad'].isin(['Bogot', 'Medelln', 'Cali']), 'Mas_2_mill'] = 'No'

# Las primeras filas del nuevo DataFrame
df.head()
````

Podemos agrupar los datos en nuestro `DataFrame` segn la variable `Mas_2_mill` y luego sumar los gastos en `Salud`, `Transporte`, y `Educacin` para cada grupo. Esto se puede hacer utilizando las funciones `groupby()` y `sum()`. Veamos cmo hacerlo.

````python
# Agrupamos los datos por 'Mas_2_mill' y sumamos los gastos de 'Salud', 'Transporte', y 'Educacin'
df_grouped = df.groupby('Mas_2_mill')[['Salud', 'Transporte', 'Educacin']].sum()

# Las primeras filas del nuevo DataFrame
df_grouped
````

#### Usando la funcin `apply`

La funcin `apply()` es una herramienta muy poderosa en `pandas` que nos permite aplicar una funcin a lo largo de los ejes de un `DataFrame`. En otras palabras, podemos aplicar una funcin a cada columna o fila de un `DataFrame`. Esto es especialmente til cuando queremos realizar operaciones de agregacin personalizadas en nuestros datos.

Por ejemplo, podemos usar la funcin `apply()` para sumar los gastos en todas las categoras para los grupos `Mas_2_mill` creados anteriormente:

````python
# Agrupamos los datos 'Mas_2_mill' y sumamos los gatos en todas las categoras usando `apply`
df_grouped_all = df.groupby('Mas_2_mill').apply(lambda x: x.iloc[:, 1:13].sum())

# Display the grouped data
df_grouped_all
````

##### Funciones lambda

Las funciones `lambda()` como la utilizada anteriormente es una pequea funcin annima. Una funcin lambda puede tomar cualquier nmero de argumentos, pero solo puede tener una expresin. Las funciones lambda se utilizan cuando necesitamos una funcin pequea y annima por un corto perodo de tiempo.

En el contexto de [pandas](https://pandas.pydata.org/), las funciones lambda se utilizan ampliamente con las funciones `apply()`, `map()`, y `filter()`. En el ejemplo anterior, utilizamos una funcin lambda dentro de la funcin `apply()` para sumar los gastos en todas las categoras para cada grupo de 'Mas_2_mill'.

La sintaxis de una funcin lambda en `Python` es la siguiente:

```python
lambda arguments: expression
```

Donde:
- `lambda` es la palabra clave que define una funcin lambda.
- `arguments` son los argumentos que se pasan a la funcin.
- `expression` es la expresin que se evala y devuelve.

## [Pandas](https://pandas.pydata.org/) para Anlisis Descriptivo

Todas las tablas siempre deben estar acompaadas de un texto que explique brevemente que contiene la tabla y resaltar estadsticas de inters. Por ejemplo, el gasto promedio mensual ms alto es `Alojamiento, agua, electricidad, gas y otros combustibles`, 167.8 mil (47.09), y el menor es en `Salud`, 12.5 mil (5.5). Sin embargo, la naturaleza del proyecto determinar la mejor manera de describir nuestros datos. Por ejemplo, si el proyecto tiene que ver con la relacin que existe entre las variables, el anlisis tiene que estar complementado con medidas que resuman esta relacin. Por ejemplo, podemos calcular la matriz de correlacin de estos datos con `corr()`

````python
# Calculamos la matriz de correlacion
correlation_matrix = df.corr()
correlation_matrix
````

y luego generar una visualizacin haciendo uso de las libreras [matplotlib](https://matplotlib.org/) y [seaborn](https://seaborn.pydata.org/) que nos facilite la descripcin:

````python
# Importamos matplotlib y seaborn para visualizar datos
import seaborn as sns
import matplotlib.pyplot as plt



# Creamos un mapa de calor
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
````

Recordemos que correlacin es una medida estadstica que describe la magnitud y la direccin de la **relacin lineal** entre dos variables. Los valores de correlacin varan entre -1 y 1. Un valor de correlacin de 1 indica una fuerte correlacin positiva, donde las dos variables tienden a aumentar juntas. Un valor de -1 indica una fuerte correlacin negativa, donde una variable tiende a aumentar cuando la otra disminuye. Un valor de 0 indica que no hay correlacin. En el mapa de calor, los colores ms clidos (ms cercanos al rojo) indican una correlacin positiva ms fuerte, mientras que los colores ms fros (ms cercanos al azul) indican una correlacin negativa ms fuerte. Los valores en cada celda representan el coeficiente de correlacin entre las dos variables correspondientes.

Podemos ver en la grfica por ejemplo que el gasto en  `Alimentos y bebidas no alcohlicas` est negativamente correlacionado con `Bebidas alcohlicas y tabaco` y que el gasto en `Muebles, artculos para el hogar y para la conservacin ordinaria del hogar` est altamente correlacionado con el gasto en `Bienes y servicios diversos`.

Otra forma de estudiar la relacin entre variables es examinar los grficos de dispersin bivariados q entre pares de variables.
